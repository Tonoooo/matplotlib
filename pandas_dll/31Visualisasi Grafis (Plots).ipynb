# membuat DataFrame baru

```python
import numpy as np
```

```python
np.random.seed(100) # agar angka randomnya sama dengan tutor
```

```python
# membuat list dari 1 - 100
angka=[]
for i in range(1,101):
    angka.append(i)
```

```python
angka
```

```python
baru = pd.DataFrame(np.random.rand(100,5),angka,['A','B','C','D','E'])
# urutan mengisi DataFrame(self, data, index, columns, dtype, copy)
# yang diisi disini DataFrame(datanya(datanya random 0 sampai 1 dengan dimensi 100 baris 5 kolom),index,kolom)

# bedanya rand dan randn: kalo (rand tanpa n) antara 0 sampai 1, jika randn (pake n) dari -1 sampai 1
```

```python
baru.head()
```

# membuat plot area

```python
# membuat plot area
baru.plot.area()
```

```python
baru.plot.area(alpha=0.5)
```

# membuat scatter plot
titik titik bulet

```python
# scatter() membutuh kan sumbu x dan sumbu y
baru.plot.scatter(x='A',y='B', c='E') 

# sumbu x nya kolom a dan sumbu y nya kolom b, c adalah color jadi warna nya berdasarkan kolom E
```

```python
baru.plot.scatter(x='A',y='B', c='E', cmap='ocean') # warnanya berdasarkan E dan warna laut
```

```python
baru.plot.scatter(x='A',y='B', c='E', cmap='Reds') # warnanya berdasarkan E dan warna merah
```

```python
# membedakan titik titik nya berdasarkan ukuran
baru.plot.scatter(x='A',y='B', s=baru.D) 
```

```python
baru.plot.scatter(x='A',y='B', s=baru.D*100) 
# jadi skala ukurannya berdasarkan kolom D
```

```python
baru.plot.scatter(x='A',y='B', s=baru.D*100, c='E', cmap='Reds') 
```

# membuat box plot

```python
baru.plot.box()
```

# membuat line plot

```python
baru.plot.line(y='B') # sumbu y adalah kolom b
```

```python
baru.plot.line(y='B',lw=0.5) # lw = linewidth = tebal baris
```

```python
# mengatur panjang dan lebar plot nya
baru.plot.line(y='B',lw=0.5, figsize=(15,5)) # panjang 15 lebar 5
```
