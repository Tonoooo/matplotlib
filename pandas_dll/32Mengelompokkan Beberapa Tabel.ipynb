```python
import pandas as pd
```

```python
satu = pd.DataFrame({'A':['a0','a1','a2','a3','a4','a5'],
        'B':['b0','b1','b2','b3','b4','b5'],
        'C':['c0','c1','c2','c3','c4','c5'],
        'D':['d0','d1','d2','d3','d4','d5']},
        index=[0,1,2,3,4,5])
```

```python
dua = pd.DataFrame({'A':['a6','a7','a8','a9','a10','a11'],
        'B':['b6','b7','b8','b9','b10','b11'],
        'C':['c6','c7','c8','c9','c10','c11'],
        'D':['d6','d7','d8','d9','d10','d11']},
        index=[6,7,8,9,10,11])
```

```python
tiga = pd.DataFrame({'A':['a12','a13','a14','a15','a16','a17'],
        'B':['b12','b13','b14','b15','b16','b17'],
        'C':['c12','c13','c14','c15','c16','c17'],
        'D':['d12','d13','d14','d15','d16','d17']},
        index=[12,13,14,15,16,17])
```

```python
satu
```

```python
dua
```

```python
tiga
```

# concatenation (penggabungkan)
# menggabungkan 3 dataframe ini menjadi satu

```python
pd.concat([satu,dua,tiga]) 
# ini default axis=0(axis: 0=baris, 1=kolom) asalkan nama kolomnya sama semua
```

```python
pd.concat([satu,dua,tiga],axis=1) # axis: 0=baris, 1=kolom
```

# menggabung kan bebrapa baris/kolom berdasarkan kesamaan

```python
# menggabung kan baris/kolom berdasarkan kesamaan
bisnis = pd.DataFrame({'Perusahaan':['telkom','telkom','pertamina','bri','pertamina','kfc','kfc','indosat'],
                          'karyawan':['joko','budi','susi','irana','bambang','burhanudin','intan','mukiman'],
                          'usia':[44,53,67,34,60,19,20,17]})
```

```python
bisnis
```

```python
bisnis.groupby('Perusahaan') # mengelompokkan kolom perusahaan
```

```python
bisnis.groupby('Perusahaan').sum()
# isi baris yang sama maka usia ditambhakan
# disini karyawan tidak ada karna string bukan numerik
```

```python
# untuk jumlah data
bisnis.groupby('Perusahaan').count()
```

```python
# rata rata
bisnis.groupby('Perusahaan').mean()
```

```python
bisnis.groupby('Perusahaan').min()
# di barisnya diurutkan paling kecil sesuai karyawan menurut abjad 
```

```python
bisnis.groupby('Perusahaan').max()
# di barisnya diurutkan paling kecil sesuai karyawan menurut abjad 
```

```python
bisnis.groupby('Perusahaan').usia.min() # urutan paling kecil disetiap barisnya
```

```python
bisnis.groupby('Perusahaan').describe()
# bri dan indosat NaN karna cuma punya 1 data saja
```

```python
# sama saja tapi dibalik/tanspose
bisnis.groupby('Perusahaan').describe().transpose() 
```

# #bentuk dataFrame yang bentuknya misip dengan sql

```python
satuu = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                      'one' : [1,2,3,4,5],
                      'two' : [6,7,8,9,10]})
```

```python
duaa = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                      'three' : [1,2,3,4,5],
                      'four' : [6,7,8,9,10]})
```

```python
satuu
```

```python
duaa
```

# menggabungkan 2 dataframe dengan menggunakan parameter key

# merge()
# inner,left,right,outer

```python
baru = pd.merge(satuu, duaa, how='inner', on='key')
# inner = gunakan keys nya dari dataframe
```

```python
baru
```

```python
ssatu = pd.DataFrame({'key1':['k0','k0','k1','k2'],
                      'key2':['k0','k1','k0','k1'],
                      'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})

ddua = pd.DataFrame({'key1':['k0','k1','k1','k2'],
                      'key2':['k0','k0','k0','k0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
ssatu
```

```python
ddua
```

```python
baru2 = pd.merge(ssatu,ddua,how='left', on=['key1','key2'])
# left= penggabungannya lebih diutamakan yang left(dataframe yang pertama) 
```

```python
baru2
```

```python
baru3 = pd.merge(ssatu,ddua,how='right', on=['key1','key2'])
# right= penggabungannya lebih diutamakan yang right(dataframe yang kedua) 
```

```python
baru3
```

```python
baru4 = pd.merge(ssatu,ddua,how='outer', on=['key1','key2'])
# outer = semua dimasukkan
```

```python
baru4
```

# digabungkan tapi keys nya ada di index bukan dibaris
# join()

```python
bapak = pd.DataFrame({'A':['A0','A1','A2'],
                      'B':['B0','B1','B2']},
                        index=['k0','k1','k2'])

ibu = pd.DataFrame({'C':['C0','C2','C3'],
                      'D':['D0','D2','D3']},
                        index=['k0','k2','k3'])
```

```python
bapak
```

```python
ibu
```

```python
bapak.join(ibu)
```

```python
bapak.join(ibu,how='inner') # menggbungkan yang key nya sama persis
```
