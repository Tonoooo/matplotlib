```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math
```

Matriks adalah susunan angka-angka dalam bentuk bujur sangkar atau persegi panjang yang diapit oleh tanda kurung. Semua bilangan yang ada dalam matriks adalah bilangan real. ... Vektor adalah matriks yang hanya memiliki satu kolom saja atau hanya memiliki satu baris saja.

```python
baris = np.array([2,3])  # untuk vektor baris
kolom = np.array([[1],[4]]) # untuk vektor kolom
```

```python
print(baris)
print(kolom)
```

```python
# menampilkan vektor 
display(Math(sym.latex(sym.simplify(baris))))
```

```python
display(Math(sym.latex(sym.simplify(kolom))))
```

```python
# membuat visualisasi
plt.plot([0, baris[0]], [0, baris[1]], 'r', label='vektor baris') # [0, baris[0]], [0, baris[1]] -> 0 artinya dimulai dari nol sampai index baris 0 dan 0 sampai index baris 1
plt.plot([0, kolom[0]], [0, kolom[1]], 'b', label='vektor kolom')

plt.axis('square') # agar siplotnya kotak
plt.axis([-5,5,-5,5]) # sumbu x nya -5 sampai 5 dan y nya -5 sampai 5
plt.grid()
plt.legend()
plt.show()
```

```python
# secara konsep vektor baris dan kolom itu berbeda tapi kalo secara geomrtis itu dua vektor yang sama
```

```python
# membuat vektor 3 dimensi
a = np.array([2,0,-3])
b = np.array([-2,2,4])
```

```python
fig = plt.figure(figsize=plt.figaspect(1))  # namanya bebas. tidak harus pake fig
ax = fig.add_subplot(111,projection='3d')  # namanya bebas. tidak harus pake ax

ax.plot([0, a[0]], [0, a[1]], [0, a[2]], 'r', linewidth=3) # linewidth = ketebalan garis
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'b', linewidth=3)

ax.plot([-5,5],[0,0],[0,0],'--',color=[.6,.6,.6]) # membuat garis imajiner
ax.plot([0,0],[-5,5],[0,0],'--',color=[.6,.6,.6])
ax.plot([0,0],[0,0],[-5,5],'--',color=[.6,.6,.6])

ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)
plt.show()
```
