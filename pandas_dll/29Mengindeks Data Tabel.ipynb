# iloc
# .iloc [baris,kolom]
# metode untuk irisan atau potongan

```python
dataku.iloc[0] # memilih sesuai indexnya. [0] data yang diindex 0, (menampilkan data yang ada di index 0)
```

```python
dataku.iloc[0:11]
```

```python
type(dataku.iloc[0:11])
```

```python
dataku.iloc[-5:] # dari index -5 sampai belakang terus. yang paling terakhir indexnya -1
```

```python
dataku.iloc[-5:-1,-4:]
# [-5:-1,-4:] = dari index -5 sampai -1, -4: untuk kolomnya

# -4: = dari index -4 sampai kebelakang terus

# kolom deck     index nya kalo dihintung dari depan adalah 8 jika dari belakang/ujung adalah -1 
# kolom embarked index nya kalo dihintung dari depan adalah 7 jika dari belakang adalah -2
# kolom fare     index nya kalo dihintung dari depan adalah 6 jika dari belakang adalah -3
# kolom parch    index nya kalo dihintung dari depan adalah 5 jika dari belakang adalah -4
```

```python
dataku.iloc[-5:-1,-4:-2] #dari index -5 sampai -1, -4:-2 untuk kolomnya dari index -4 sampai index -2
```

```python
dataku.iloc[[0,2,4]] # menampilkan hanya baris index ke 0,2,4
```

```python
dataku.iloc[[0,2,4],[1,3,6]] # menampilkan hanya baris index ke 0,2,4 dan kolom index ke 1,3,6
```

# loc
# .loc['nama baris','nama kolom']

```python
# loc = jika tidak tahu index tapi hanya tau nama baris/nama kolom
```

```python
# menganti index baris nya jadi string
databaru = pd.read_csv('kapal_titanic.csv', index_col='embarked')
#index barisnya sesuai dengan kolom embarked
```

```python
databaru
```

```python
databaru.loc['S'] # semua baris yang berlabel S 
```

```python
databaru.loc['S','age'] #loc tidak harus berupa list
```

```python
# menampilkan irasan lebih dari satu kolom
databaru.loc['S',['age','fare']]
```

```python
# menampilkan irasan lebih dari satu kolom dan baris
databaru.loc[['S','Q'],['age','fare']]
```
